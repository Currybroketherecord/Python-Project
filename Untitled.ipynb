```python
import pandas as pd

# Load dataset
df = pd.read_csv("country_wise_latest.csv")

# Check column names
print(df.columns)

# Preview first 5 rows
print(df.head())

# Identify missing values
print(df.isnull().sum())
```

```python
# Convert 'Confirmed', 'Deaths', 'Recovered', and other numeric columns to proper format
numeric_cols = ["Confirmed", "Deaths", "Recovered", "Active", "New cases", "New deaths", "New recovered"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Calculate Death Rate & Recovery Rate
df["Death Rate (%)"] = (df["Deaths"] / df["Confirmed"]) * 100
df["Recovery Rate (%)"] = (df["Recovered"] / df["Confirmed"]) * 100

# Handle missing values (fill or interpolate)
df.fillna(method='ffill', inplace=True)  # Forward fill missing values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 countries by total confirmed cases
top_countries = df.sort_values("Confirmed", ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_countries["Country/Region"], x=top_countries["Confirmed"])
plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.grid()
plt.show()

# Death Rate vs. Recovery Rate Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Death Rate (%)", y="Recovery Rate (%)", hue="WHO Region")
plt.title("COVID-19 Death vs. Recovery Rates by Country")
plt.xlabel("Death Rate (%)")
plt.ylabel("Recovery Rate (%)")
plt.grid()
plt.show()
```

```python
# Compare case trends for selected countries
selected_countries = ["USA", "India", "Kenya", "Brazil"]
df_selected = df[df["Country/Region"].isin(selected_countries)]

plt.figure(figsize=(10, 5))
sns.barplot(x="Country/Region", y="Confirmed", data=df_selected)
plt.title("Total COVID-19 Cases Comparison")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.grid()
plt.show()
```
